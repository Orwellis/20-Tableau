```python
import pandas as pd
import os


```

```python
x = "Total Annual Memberships Sold"
print("Annual Member" in x and "all time" not in x.lower())
```

```python
data_frames = []

def rename_columns(df):
    annual_member_column = None
    day_passes_column = None
    
    for column in list(df.columns):
        if "Annual Member" in column and "Sign-Ups" not in column:
            annual_member_column = column
            df = df.rename(columns={annual_member_column: "Annual Memberships Sold"})
        if "24-Hour" in column:
            day_passes_column = column
            df = df.rename(columns={day_passes_column: "24-Hour Passes Purchased (midnight to 11:59 pm)"})
    return df
        

for subdir, dirs, files in os.walk('./'):
    for file in files:
        if ".csv" in file:
            data = pd.read_csv(file)
            updated_data = rename_columns(data)
            data_frames.append(updated_data)
            
data_frames[0]
```

```python
columns = []
for df in data_frames:
    columns += list(df.columns)
    
possible_columns = set(columns)

column_count = {}

for column in possible_columns:
    column_count[column] = 0

for df in data_frames:
    for column in list(df.columns):
        column_count[column] += 1
        
print(column_count)


name = []
count = []

for key in column_count:
    name.append(key)
    count.append(column_count[key])
    
count_df = pd.DataFrame({"name":name, "count":count})
count_df.sort_values("count")


shared_columns = list(count_df[count_df["count"] == 23]["name"])
```

```python
reduced_dfs = [df[shared_columns] for df in data_frames]
merged_df = pd.concat(reduced_dfs)
merged_df = merged_df.sort_values("Date")
merged_df.head()
```

```python
merged_df.to_csv("./../merged_daily_data.csv")
```
